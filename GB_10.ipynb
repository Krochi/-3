```python
#Задача1:
import pandas as pd
import random

lst = ['robot'] * 10 + ['human'] * 10
random.shuffle(lst)

data = pd.DataFrame({'whoAmI': lst})

unique_values = data['whoAmI'].unique()

one_hot_df = pd.DataFrame()

for value in unique_values:
    one_hot_df[value] = (data['whoAmI'] == value).astype(int)

one_hot_df.head()
```

```python
#Задача2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'age': [25, 34, 22, 45, 36, 52, 28, 40, 33, 47],
    'spending_score': [75, 65, 80, 55, 60, 45, 70, 50, 68, 40]
}

df = pd.DataFrame(data)

df.info()
df.head()

plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='spending_score', data=df, marker="o")

plt.title('Взаимосвязь возраста и расходов', fontsize=16)
plt.xlabel('Возраст (Age)', fontsize=14)
plt.ylabel('Балл по расходам (Spending Score)', fontsize=14)
plt.grid(True)

plt.show()
```

```python
# Задача3
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = {
    'age': [25, 34, 22, 45, 36, 52, 28, 40, 33, 47],
    'spending_score': [75, 65, 80, 55, 60, 45, 70, 50, 68, 40]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.lineplot(x='age', y='spending_score', data=df, marker="o", color='b', linewidth=2)

plt.title('Влияние возраста на расходы сотрудников', fontsize=20, color='darkblue')
plt.xlabel('Возраст (лет)', fontsize=14, color='darkgreen')
plt.ylabel('Балл по расходам', fontsize=14, color='darkgreen')

plt.xticks(fontsize=12, color='black')
plt.yticks(fontsize=12, color='black')
plt.grid(True, which='both', linestyle='--', linewidth=0.7)

plt.show()
```
